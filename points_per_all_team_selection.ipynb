```python
import pandas as pd
import numpy as np
```

```python
#Open and read dataset
path1 = open('./Datasets/awards_data.csv')
path2 = open('./Datasets/player_stats.csv')
awards = pd.read_csv('./Datasets/awards_data.csv')
player_data = pd.read_csv('./Datasets/player_stats.csv')
#Call head() to make sure correct dataset was selected
awards.head()
```

```python
player_data.head()
```

```python
#merging the files
awarded_player_stats = awards[["season","nbapersonid", "All NBA First Team", "All NBA Second Team", "All NBA Third Team", "all_star_game"]].merge(player_data[["nbapersonid", "games","points"]],on="nbapersonid", how= "right")
```

```python
awarded_player_stats.to_excel("stats.xlsx", index = False)
```

```python
#Replacing all NAN with 0
new_data = awarded_player_stats.fillna(0)
```

```python
#Viewing the two datasets merged
new_data.head()
```

```python
#Selecting rows of players who won All Nba First Team in seasons from 2007 to 2021 
awards1 = new_data[(new_data['season'] > 2006) & (new_data['season'] < 2022) & (new_data['All NBA First Team'] > 0.0)]
```

```python
ppg = awards1['points'].sum()/awards1['games'].sum()
rounded_ppg = (round(ppg, 1))
```

```python
#Selecting rows of players who won All Nba Second Team in seasons from 2007 to 2021 
awards2 = new_data[(new_data['season'] > 2006) & (new_data['season'] < 2022) & (new_data['All NBA Second Team'] > 0.0)]
```

```python
ppg2 = awards2['points'].sum()/awards2['games'].sum()
rounded_ppg2 = round(ppg2, 1)
```

```python
#Selecting rows of players who won All Nba Third Team in seasons from 2007 to 2021 
awards3 = new_data[(new_data['season'] > 2006) & (new_data['season'] < 2022) & (new_data['All NBA Third Team'] > 0.0)]
```

```python
ppg3 = awards3['points'].sum()/awards3['games'].sum()
rounded_ppg3 = round(ppg3, 1)
```

```python
#Selecting rows of players who were in the all star game in seasons from 2007 to 2021 
awardsStar = new_data[(new_data['season'] > 2006) & (new_data['season'] < 2022) & (new_data['all_star_game'] > 0.0)]
```

```python
ppg4 = awardsStar['points'].sum()/awardsStar['games'].sum()
rounded_ppg4 = round(ppg4, 1)
```

```python
print(f"1st Team: {rounded_ppg} points per game")
print(f"2nd Team: {rounded_ppg2} points per game")  
print(f"3rd Team: {rounded_ppg3} points per game")  
print(f"All-Star: {rounded_ppg4} points per game")  
```
