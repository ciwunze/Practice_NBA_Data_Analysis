```python
import pandas as pd
import numpy as np
```

```python
#Open and read dataset
path1 = open('./Datasets/awards_data.csv')
path2 = open('./Datasets/player_stats.csv')
awards = pd.read_csv('./Datasets/awards_data.csv')
player_data = pd.read_csv('./Datasets/player_stats.csv')
#Call head() to make sure correct dataset was selected
awards.head()
```

```python
player_data.head()
```

```python
#merging the files
awarded_player_stats = awards[["season","nbapersonid", "All NBA First Team", "All NBA Second Team", "All NBA Third Team"]].merge(player_data[["nbapersonid", "player", "draftyear"]],on="nbapersonid", how= "left")
```

```python
awarded_player_stats.head()
```

```python
#Replacing all NAN with 0
new_data = awarded_player_stats.fillna(0)
```

```python
#Viewing the two datasets merged
new_data.head()
```

```python
#Narrow group of players drafted 2007 and onwards, who made All NBA Selection
allTeam = new_data[(new_data['season'] > 2006) & (new_data['draftyear'] > 2006) & ((new_data['All NBA First Team'] > 0.0)|(new_data['All NBA Second Team'] > 0.0)|(new_data['All NBA Third Team'] > 0.0))]
```

```python
allTeam
```

```python
#Remove duplicates and use the player's first appearance in All NBA Selection
new = allTeam.groupby(by=['nbapersonid']).first()
new.head()
```

```python
#Create empty list to input values
num = []
#Iterate through Series to work through calculations
for index, row in new.iterrows():
    num.append(row['season'] - row['draftyear'] + 1)
num_of_years = round(sum(num)/ len(num),1)
print(num_of_years)
print(f"The average number of years of experience it takes to make it first All NBA Selection is {num_of_years} years.")
```
